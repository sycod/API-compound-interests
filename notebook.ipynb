# Initialisation

```python
import yaml
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
# CONFIG
# Load config file
with open("config.yaml", "r", encoding="utf-8") as f:
    config = yaml.safe_load(f)
```

```python
# Setup
k = config["k"]["init"]
rate = config["rate"]["init"]
ann_savings = config["ann_savings"]["init"]
DURATION = config["duration_yrs"]
TAX = config["tax"]
```

```python
# üöß
rate = 0.1 ; ann_savings = 1000
```

```python
df = pd.DataFrame(columns=["annee", "epargne", "brut", "epargne_cum", "comp_brut", "capital", "cum_net", "mensuel_net"])
```

# Ann√©es

```python
df["annee"] = range(DURATION)
df["annee"] = df["annee"] + 1
```

# √âpargne cumul√©e

```python
df["epargne"] = k + ((df["annee"] - 1) * ann_savings)
```

# Int√©r√™ts annuels bruts

Chaque nouvelle ann√©e commence avec le **capital de d√©part** + l'**√©pargne** r√©alis√©e dans l'an√©e + les **int√©r√™ts de l'ann√©e pass√©e**

Ce sont ces int√©r√™ts qui, r√©inject√©s dans le calcul des int√©r√™ts de l'ann√©e en cours, forment les **int√©r√™ts compos√©s**.

```python
brut = [k * rate]
for i in range(1, len(df.index)):
    

    n = brut[i-1]
    brut.append(n)


#     cum = df.brut.loc[:i-1].cumsum()[i-1]
df["brut"] = brut
df["brut"] = round(df["brut"], 0).astype(int)
```

```python
display(f"{k = }, {rate = }, {ann_savings = }, {DURATION = }, {TAX = }")
display(df)
```

# Capital

```python
df["capital"] = df["comp_brut"] + df["epargne_cum"]
```

# Int√©r√™ts cumul√©s nets annuels

```python
df["cum_net"] = round(df["cum_brut"] * (1 - TAX), 0).astype(int)
```

# Int√©r√™ts nets mensuels

```python
df["mensuel_net"] = round(df["cum_net"] / 12, 0).astype(int)
```
