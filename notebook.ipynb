# Initialisation

```python
import yaml
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
# CONFIG
# Load config file
with open("config.yaml", "r", encoding="utf-8") as f:
    config = yaml.safe_load(f)
```

```python
# Setup
k = config["k"]["init"]
rate = config["rate"]["init"]
ann_savings = config["ann_savings"]["init"]
DURATION = config["duration_yrs"]
TAX = config["tax"]
```

```python
# üöß
rate = 0.1 ; ann_savings = 1000
```

```python
df = pd.DataFrame(columns=["annee", "epargne_cum", "brut", "cum_brut", "comp_brut", "capital", "cum_net", "mensuel_net"])
```

# Ann√©es

```python
df["annee"] = range(DURATION)
df["annee"] = df["annee"] + 1
```

# √âpargne cumul√©e

```python
df["epargne_cum"] = k + ((df["annee"] - 1) * ann_savings)
```

# Int√©r√™ts annuels bruts

```python
df["brut"] = round((df["epargne_cum"]) * (rate), 0).astype(int)
```

# Int√©r√™ts cumul√©s bruts

```python
df["cum_brut"] = df["brut"].cumsum()
```

# Int√©r√™ts compos√©s bruts

```python
df["comp_brut"] = df["cum_brut"]
```

```python
display(f"{k = }, {rate = }, {ann_savings = }, {DURATION = }, {TAX = }")
display(df)
```

# Capital

```python
df["capital"] = df["comp_brut"] + df["epargne_cum"]
```

# Int√©r√™ts cumul√©s nets annuels

```python
df["cum_net"] = round(df["cum_brut"] * (1 - TAX), 0).astype(int)
```

# Int√©r√™ts nets mensuels

```python
df["mensuel_net"] = round(df["cum_net"] / 12, 0).astype(int)
```
